```python
import tensorflow as tf
from tensorflow import keras
from keras import Input,layers,Model
from matplotlib import pyplot as plt

#Convert to tensorflow lite model
import os
import h5py
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from sklearn.metrics import accuracy_score
from sys import getsizeof

def get_file_size(path):
  size = os.path.getsize(path)
  return size

def convert_bytes(size, unit= None):
  if unit=="KB":
    return print("File size : " + str(round(size/1204,3)) + 'Kilobytes')
  elif unit=="MB":
    return print("File size: " + str(round(size/1024*1024,3)) + 'Megabytes')
  else:
    return print("File size: "+ str(round(size)) + 'Bytes')



mnist = keras.datasets.mnist
(train_images,train_labels), (test_images,test_labels) = mnist.load_data()

train_images = train_images/255
test_images = test_images/255
#X_train = keras.utils.normalize(X_train(X_train,axis = 1))
#Y_train = keras.utils.normalize(Y_train,axis = 1)
plt.imshow(train_images[0], interpolation = 'nearest')
plt.show()
train_images = train_images.reshape(len(train_images), 28*28)
test_images = test_images.reshape(len(test_images), 28*28)
print(train_images)
print(test_images)

#seq_model.fit(X_train, Y_train, epochs = 10)
#val_loss , val_accuracy = seq_model.evaluate(test_images, test_images)
#print(val_loss)
#print(val_accuracy)



```

```python

import numpy as np
import matplotlib.pyplot as plt
class_names = ['one','Two','Three','Four','Five','Six','Seven','Eight','Nine']
#Dimension of train images
print(train_images.shape)
print(len(train_labels))
print(np.unique(train_labels))

#Dimension of test image
print(test_images.shape)
print(len(test_images))

#Plotting a sample image
plt.figure()
plt.imshow(train_images[18].reshape(28,28))
#plt.colorbar()
#plt.grid(False)
#plt.show()
print(class_names[train_labels[18]])

#Normalizing
train_images = train_images/255.0

import numpy as np
import matplotlib.pyplot as plt
class_names = ['one','Two','Three','Four','Five','Six','Seven','Eight','Nine']
#Dimension of train images
print(train_images.shape)
print(len(train_labels))
print(np.unique(train_labels))

#Dimension of test image
print(test_images.shape)
print(len(test_images))

#Plotting a sample image
plt.figure()
#plt.imshow(train_images[18])
#plt.colorbar()
#plt.grid(False)
plt.show()
print(class_names[train_labels[18]])

#Normalizing
train_images = train_images/255.0
test_images = test_images/255.0

#Building model
model = keras.Sequential([keras.layers.Flatten(input_shape = (28,28)),
                          keras.layers.Dense (128, activation = 'relu'),
                          keras.layers.Dense (10, activation = 'softmax')])
#Compiling the model
model.compile(optimizer = 'adam',
              loss = keras.losses.SparseCategoricalCrossentropy (from_logits = True),
              metrics = ['accuracy'])

#Training the model  model.fit(train_images,train_labels ,epochs=10)

#Saving the model
Keras_Model_Name = "tf_model_fashion_mnist.h5"
model.save (Keras_Model_Name)
convert_bytes (get_file_size (Keras_Model_Name), "MB")

test_images = test_images.reshape(len(test_images), 28, 28)

test_loss, test_acc = model.evaluate (test_images, test_labels, verbose = 2)
print("\nTest accuracy: ", test_acc)

#Converting the tf model into tflite model
tf_lite_converter = tf.lite.TFLiteConverter.from_keras_model(model)
tf_lite_converter_optimizations = [tf.lite.Optimize.DEFAULT]

#Writing the tflite model in colab repository
tflite_model = tf_lite_converter.convert()
tflite_model_name = "tf_lite_model"
open(tflite_model_name, "wb").write (tflite_model)

#Converting the file size in kb
#TF_LITE_FILE_NAME = tflite_model_name
convert_bytes (get_file_size(tflite_model_name), "KB")

#Compiling the model
model.compile(optimizer = 'adam',
              loss = SparseCategoricalCrossentropy (from_logits = True),
              metrics = ['accuracy'])

#Training the model  model.fit(train_images,train_labels ,epochs=10)

#Saving the model
Keras_Model_Name = "tf_model_fashion_mnist.h5"
model.save (Keras_Model_Name)
convert_bytes (get_file_size (Keras_Model_Name), "MB")
test_loss, test_acc = model.evaluate (test_images, test_labels, verbose = 2)
print("\nTest accuracy: ", test_acc)

#Converting the tf model into tflite model
tf_lite_converter = tf.lite.TFLiteConverter.from_keras_model(model)
tf_lite_converter_optimizations = [tf.lite.Optimize.DEFAULT]

#Writing the tflite model in colab repository
tflite_model = tf_lite_converter.convert()
tflite_model_name = "tf_lite_model"
open("tflite_model_name", "wb").write (tflite_model)

#Converting the file size in kb
#TF_LITE_FILE_NAME = tflite_model_name
convert_bytes (get_file_size("tflite_model_name"), "KB")
```
